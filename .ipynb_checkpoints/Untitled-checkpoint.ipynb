### Vectorizing all the documents


```python
fields = ['section', 'question', 'text']
```

```python
matrices = {}
vectorizers = {}

for f in fields:
    cv = TfidfVectorizer(stop_words='english', min_df=5)
    X = cv.fit_transform(df[f])
    matrices[f] = X
    vectorizers[f] = cv
```

```python
matrices
```

```python
vectorizers
```

#### Here we are using similarity accross all the fields by adding the similarity in each field to get final similarity

```python
score = np.zeros(len(df))
query = "I just singned up. Is it too late to join the course?"
boost = {'question': 3.0}
for f in fields:
    b = boost.get(f, 1.0)
    q = vectorizers[f].transform([query])
    f_score = cosine_similarity(matrices[f], q).flatten()
    score = score +  f_score * b
```

```python
idx = np.argsort(score)[-5:]
```

```python
df.iloc[idx]
```

```python
filters = {
    'course': 'data-engineering-zoomcamp'
}

for field, value in filters.items():
    mask = (df[field] == value).values
    score = score * mask
```

```python
idx = np.argsort(-score)[:10]
df.iloc[idx]
```

```python
results = df.iloc[idx]
results.to_dict(orient='records')
```

## Putting All Together

```python
class TextSearch:

    def __init__(self, text_fields):
        self.text_fields = text_fields
        self.matrices = {}
        self.vectorizers = {}

    def fit(self, records, vectorizer_params={}):
        self.df = pd.DataFrame(records)

        for f in self.text_fields:
            cv = TfidfVectorizer(**vectorizer_params)
            X = cv.fit_transform(self.df[f])
            self.matrices[f] = X
            self.vectorizers[f] = cv

    def search(self, query, n_results=10, boost={}, filters={}):
        score = np.zeros(len(self.df))

        for f in self.text_fields:
            b = boost.get(f, 1.0)
            q = self.vectorizers[f].transform([query])
            s = cosine_similarity(self.matrices[f], q).flatten()
            score = score + b * s

        for field, value in filters.items():
            mask = (self.df[field] == value).values
            score = score * mask

        idx = np.argsort(-score)[:n_results]
        results = self.df.iloc[idx]
        return results.to_dict(orient='records')
```

```python
index = TextSearch(
    text_fields=['section', 'question', 'text']
)
index.fit(documents)

index.search(
    query='I just singned up. Is it too late to join the course?',
    n_results=5,
    boost={'question': 3.0},
    filters={'course': 'data-engineering-zoomcamp'}
)
```

## Vector Search
Let's use the vectorizer for the "text" field and turn it into embeddings



```python
from sklearn.decomposition import TruncatedSVD

X = matrices['text']
cv = vectorizers['text']
```

Applying dimensionality reduction, we have dense representation of document now called embedding.
SVD captures most of the information by combining synonyms(similar meaning words)

```python
svd = TruncatedSVD(n_components=16)
X_emb = svd.fit_transform(X)

X_emb[0]
```

```python
query = 'I just singned up. Is it too late to join the course?'

Q = cv.transform([query])
Q_emb = svd.transform(Q)
Q_emb[0]
```

```python
np.dot(X_emb[0], Q_emb[0])
```

```python
score = cosine_similarity(X_emb, Q_emb).flatten()
idx = np.argsort(-score)[:10]
list(df.loc[idx].text)
```

## Non-Negative Matrix Factorization


```python
from sklearn.decomposition import NMF
nmf = NMF(n_components=16)
X_emb = nmf.fit_transform(X)
X_emb[0]
```

```python
Q = cv.transform([query])
Q_emb = nmf.transform(Q)
Q_emb[0]
```

```python
score = cosine_similarity(X_emb, Q_emb).flatten()
idx = np.argsort(-score)[:10]
list(df.loc[idx].text)
```

## Bidirectional Encoder Representations from Transformers (BERT)

```python
import torch
from transformers import BertModel, BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")
model.eval()  # Set the model to evaluation mode if not training
```

```python
#tokenizing
texts = [
    "Yes, we will keep all the materials after the course finishes.",
    "You can follow the course at your own pace after it finishes"
]
encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
```

```python
#computing embeddings
with torch.no_grad():  # Disable gradient calculation for inference
    outputs = model(**encoded_input)
    hidden_states = outputs.last_hidden_state
```

```python
sentence_embeddings = hidden_states.mean(dim=1)
sentence_embeddings.shape
```

```python
X_emb = sentence_embeddings.numpy()
```

```python
sentence_embeddings_cpu = sentence_embeddings.cpu()
```

```python
def make_batches(seq, n):
    result = []
    for i in range(0, len(seq), n):
        batch = seq[i:i+n]
        result.append(batch)
    return result
```
